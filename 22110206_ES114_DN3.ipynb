```python
#How do the total points won by each player compare across different rounds of the tournament?
import pandas as pd
import matplotlib.pyplot as plt


data = pd.read_csv('/content/AusOpen-men-2013.csv')

plt.scatter(data['Round'], data['TPW.1'], label='Player1')
plt.scatter(data['Round'], data['TPW.2'], label='Player2')

plt.xlabel('Round')
plt.ylabel('Total Points Won')
plt.title('Total Points Won by Players in Different Rounds')

plt.legend()

plt.show()
```

```python
#Is there a correlation between the number of net points attempted by Player 2 and their second serve percentage in each match?
import pandas as pd
import seaborn as sns

# Load data from CSV file
tennis_data = pd.read_csv("/content/FrenchOpen-men-2013.csv")

# Create a bar plot using Seaborn
sns.barplot(x="NPA.2", y="SSP.2", data=tennis_data)

plt.xlabel("Net Points Attempted by Player 2")
plt.ylabel("Second Serve Percentage for Player 2")
plt.title("Relationship between Net Points Attempted and Second Serve Percentage for Player 2")

plt.show()
```

```python
#How does the net points won (NPW) by player 1 compare to that of player 2 in the tournament?
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('/content/USOpen-men-2013.csv')
plt.figure(figsize=(10, 8))

sns.boxplot(x='NPW.1', y='NPW.2', data=df)


plt.title('Comparison of Net Points Won by Players 1 and 2')
plt.xlabel('Net Points Won by Player 1')
plt.ylabel('Net Points Won by Player 2')

xticks = plt.xticks()[0]
plt.xticks(xticks, xticks.astype(int))
plt.show()
```

```python
#What is the percentage distribution of the different types of results (win/loss) for Player 1 in the tournament?
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('/content/USOpen-women-2013.csv')
wins = data['Result'].value_counts()[1]
losses = data['Result'].value_counts()[0]

# Calculate the percentage distribution
win_percentage = (wins / (wins + losses)) * 100
loss_percentage = (losses / (wins + losses)) * 100

# Create a pie chart
plt.figure(figsize=(6,6))
labels = ['Wins', 'Losses']
sizes = [win_percentage, loss_percentage]
colors = ['lightskyblue', 'lightcoral']
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.title('Percentage Distribution of Wins and Losses for Player 1')
plt.axis('equal')
plt.show()
```

```python
#How does the number of aces won by Player 1 compare to their unforced errors committed?
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv("/content/Wimbledon-men-2013.csv")

df = pd.DataFrame({'Aces Won': data['ACE.1'], 'Unforced Errors': data['UFE.1']})
sns.heatmap(df, cmap='coolwarm', fmt='g')


plt.title("Aces Won vs Unforced Errors by Player 1")
plt.xlabel("Unforced Errors")
plt.ylabel("Aces Won")
plt.show()
```

```python
#Which player had a higher net points won percentage (NPW%)?
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Read in the data from the CSV file
df = pd.read_csv('/content/Wimbledon-women-2013.csv')

# Calculate the NPW% for each player
df['NPW%_1'] = df['NPW.1'] / df['NPA.1']
df['NPW%_2'] = df['NPW.2'] / df['NPA.2']

sns.barplot(x='Player1', y='NPW%_1', data=df)
sns.barplot(x='Player2', y='NPW%_2', data=df)
plt.xticks(rotation=90)
plt.subplots_adjust(bottom=0.25)
plt.rcParams['figure.figsize'] = [20, 15]
plt.title("Net Points Won Percentage (NPW%) for Wimbledon Women's Singles Players (2013)")
plt.show()
```
